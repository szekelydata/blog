```python
import pandas as pd
```

```python
df=pd.read_json('hun2.json')
```

```python
df=df[['HARGHITA','COVASNA','MURES','BRASOV','NEAMT']].dropna(how='all') 
```

```python
conv={'HARGHITA':'HR','COVASNA':'CV','MURES':'MS','BRASOV':'BV','NEAMT':'NT'}
```

```python
conv2={}
for c in df.columns:
    print(conv[c])
    if conv[c] not in conv2:conv2[conv[c]]=[]
    d=df[[c]].dropna()
    for i in d.index:
        print(d.loc[i].dropna()[0])
        conv2[conv[c]].append(d.loc[i].dropna()[0])
    conv2[conv[c]]=sorted(conv2[conv[c]])
```

```python
s=[i+'-'+j for i in conv2 for j in conv2[i]]
```

```python
for i in s:
    print(i)
```

```python
s=[j for i in conv2 for j in conv2[i]]
```

```python
for i in s:
    print(i)
```
